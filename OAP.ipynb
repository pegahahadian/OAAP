```python
import torch
import torch.nn as nn
import torch.nn.functional as F


class HousePriceNN(nn.Module):
    def __init__(self, input_size):
        super(HousePriceNN, self).__init__()

        self.fc1 = nn.Linear(input_size, 128)
        self.relu1 = nn.ReLU()  # ReLU after first layer
        self.fc2 = nn.Linear(128, 64)
        self.relu2 = nn.ReLU()  # ReLU after second layer
        self.fc3 = nn.Linear(64, 32)
        self.relu3 = nn.ReLU()  # ReLU after third layer
        self.fc4 = nn.Linear(32, 1)  # Output layer for regression (no activation)

    def forward(self, x):
        x = torch.flatten(x, 1) 
        x = self.relu1(self.fc1(x)) 
        x = self.relu2(self.fc2(x)) 
        x = self.relu3(self.fc3(x)) 
        x = self.fc4(x)             
        return x


model_prediction = HousePriceNN(input_size=42)
```

```python
import torch
from transformers import pipeline

model_id = "meta-llama/Llama-3.2-1B"

pipe = pipeline(
    "text-generation",
    model=model_id,
    torch_dtype=torch.bfloat16,
    device_map="auto"
)

pipe("The key to life is")
```

```python
def llm(prompt, stop=["\n"]):
    response = pipe(
        prompt,
        max_length=10000,    
        do_sample=False,    
        top_p=1,             
        return_full_text=False
    )

    text_output = response[0]["generated_text"]

    for stop_token in stop:
        text_output = text_output.split(stop_token)[0]

    return text_output
```

```python
import numpy as np
import torch
from sklearn.preprocessing import StandardScaler

# Subset features and fit the scaler
selected_features = ['SIDE', 'V00CFWDTH', 'P01BMI', 'V00AGE', 'P02SEX', 'V00MCMJSW']
scaler = StandardScaler()
scaler.fit(X_train[selected_features])  # Fit on only the selected features

class MultiModelEnv:
    def __init__(self, model_prediction):
        self.model_prediction = model_prediction

    def step(self, action):
        if 'predict' in action:
            input_array = self.extract_features(action)
            input_array_trans = scaler.transform(input_array.reshape(1, -1))
            input_tensor = torch.tensor(input_array_trans, dtype=torch.float32).unsqueeze(0)
            z = self.model_prediction(input_tensor)
            self.last_input_tensor = input_tensor  # Store the input tensor for explain/evaluate
            predicted_class = torch.round(torch.sigmoid(z)).item()
            return f"The OA progression class is {int(predicted_class)} (0 = Non Progressor, 1 = Progressor)."

        elif 'finish' in action:
            return "Process complete."

        elif 'explain' in action:
            if not hasattr(self, 'last_input_tensor'):
                raise ValueError("The 'predict' action must be called before 'explain'.")

            variables = selected_features
            relevance_scores = [np.random.random() for _ in variables]  # Placeholder relevance scores
            relevance_texts = [f"{variable} has relevance score {score:.4f}" for variable, score in zip(variables, relevance_scores)]
            return "\n".join(relevance_texts)

        elif 'evaluate' in action:
            if not hasattr(self, 'last_input_tensor'):
                raise ValueError("The 'predict' action must be called before 'evaluate'.")

            noise = torch.normal(mean=0, std=0.1, size=self.last_input_tensor.shape)
            noisy_tensor = self.last_input_tensor + noise

            variables = selected_features
            sensitivity_scores = [np.random.random() for _ in variables]  # Placeholder sensitivity scores
            sensitivity_texts = [f"Sensitivity of {variable}: {sensitivity:.4f}" for variable, sensitivity in zip(variables, sensitivity_scores)]
            return "\n".join(sensitivity_texts)

        else:
            raise ValueError("Unrecognized action type")

    def extract_features(self, action):
        """
        Extract feature values from the action string.
        Assumes action string has variable=value pairs for the selected features.
        """
        import re
        variables = selected_features
        values = []
        for var in variables:
            match = re.search(rf"{var} = (\d+\.?\d*)", action)  # Match numeric values
            if match:
                values.append(float(match.group(1)))

        if len(values) != len(variables):
            raise ValueError("Missing or incorrect feature values in action string.")

        input_array = np.array(values, dtype=np.float64)
        return input_array
```
